```python
import torch
from PIL import Image
from torchvision import transforms
import matplotlib.pyplot as plt
```

```python
def resize_image(img):
    img_T = transforms.ToTensor()(img)
    resize_img = transforms.Resize(314)(img_T)
    return transforms.ToPILImage()(resize_img)

def centerCrop_image(img):
    img_T = transforms.ToTensor()(img)
    centerCrop_img = transforms.CenterCrop(299)(img_T)
    return transforms.ToPILImage()(centerCrop_img)

def randomHFlip_image(img):
    img_T = transforms.ToTensor()(img)
    randomHorizontalFlip_img = transforms.RandomHorizontalFlip()(img_T)
    return transforms.ToPILImage()(randomHorizontalFlip_img)

def randomRot_image(img):
    img_T = transforms.ToTensor()(img)
    randomRotation_img = transforms.RandomRotation(10)(img_T)
    return transforms.ToPILImage()(randomRotation_img)

def normalize_image(img):
    img_T = transforms.ToTensor()(img)
    normalize_img = transforms.Normalize([0.485, 0.456, 0.406],
                                        [0.229, 0.224, 0.225])(img_T)
    return transforms.ToPILImage()(normalize_img)

def plotFaces(imgs):
    plt.subplot2grid((2, 2), (0, 0))
    plt.imshow(imgs[0])
    plt.subplot2grid((2, 2), (0, 1))
    plt.imshow(imgs[1])
    plt.subplot2grid((2, 2), (1, 0))
    plt.imshow(imgs[2])
    plt.subplot2grid((2, 2), (1, 1))
    plt.imshow(imgs[3])
    plt.show()
```

```python
img_paths = ['./DATABASE/RAW DATA/07_thanos/07_thanos_1_f_M_34_98.tif', './DATABASE/RAW DATA/07_thanos/07_thanos_2_f_M_34_98_022.tif', './DATABASE/RAW DATA/07_thanos/07_thanos_3_f_M_34_98_020.tif', './DATABASE/RAW DATA/07_thanos/07_thanos_4_f_M_34_98.tif']
base_imgs = []
for i in img_paths:
    base_imgs.append(Image.open(i))
plotFaces(base_imgs)
img_jpegs = []
for i in base_imgs:
    img_jpegs.append(i.convert('RGB'))

plotFaces(img_jpegs)
resize_imgs = [resize_image(i) for i in img_jpegs]
plotFaces(resize_imgs)
centerCrop_images = [centerCrop_image(i) for i in resize_imgs]
plotFaces(centerCrop_images)
randomHFlip_images = [randomHFlip_image(i) for i in centerCrop_images]
plotFaces(randomHFlip_images)
randomRot_images = [randomRot_image(i) for i in randomHFlip_images]
plotFaces(randomRot_images)
normalize_images = [normalize_image(i) for i in randomRot_images]
plotFaces(normalize_images)
```
